```python
import pandas as pd

def nullcheck(file):
    print("null 개수 : ",file.isnull().sum())

def filediv(file):
    df1 =file.iloc[::2,:].reset_index(drop=True)
    df2 =file.iloc[1::2,:].reset_index(drop=True)
    
    for index,value in df2.iterrows():
        value["result"]=value["result"].split('/')[-1]
        value["result"]=value["result"].replace("_"," ")
        value["result"]=value["result"].replace(".wav.16kHz.mono.wav.txt","")
        value["result"]=value["result"].replace(".wav.txt","")
    
    df2.rename(columns={'result':'file'},inplace=True)
    
    df=pd.concat([df1,df2],axis=1)
    print(df)
    count=0
    for index,value in df.iterrows():
        if value["result"]==value["file"]:
            print(value["result"])
            count+=1
    print("맞은 개수:",count)
    print("정확률:",count/len(df)*100)
    
if __name__ =="__main__":
    filename=input()
    file=pd.read_csv(filename)
    nullcheck(file)
    filediv(file)
```
