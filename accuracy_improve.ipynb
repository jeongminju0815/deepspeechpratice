```python
a = []    # 빈 리스트 생성
 
for i in range(3):
    line = []              # 안쪽 리스트로 사용할 빈 리스트 생성
    for j in range(2):
        line.append(0)     # 안쪽 리스트에 0 추가
    a.append(line)         # 전체 리스트에 안쪽 리스트를 추가
 
print(a)
```

```python
#어간추출
line1=[]
for word1 in tokenword:
    line=[]
    for word in word1:
        stem1=stemmer.stem(word)
        line.append(stem1)
    line1.append(line)
print(line1)
```

```python
# 원형추출
wlem = nltk.WordNetLemmatizer()
line3=[]
for word1 in tokenword:
    line4=[]
    for word in word1:
        lema1=wlem.lemmatize(word)
        line4.append(lema1)
    line3.append(line4)
print(line3)
```

```python
#idea : 원형 추출한 리스트를 다시 문장을 붙이고 정답 문제도 원형을 추출한후 다시 문장으로 붙여서 비교한다.
```

```python
checklist=[]
for word in line3:
    newword="".join(word)
    print(newword)
    checklist.append(newword)
```
