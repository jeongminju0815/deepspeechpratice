```python
dic={"'s":['is','s'],"'m":['am'],"'re":['are'],"'ll":["will"],"'t":['not'],"'d":['would']}
```

```python
count2=0
for word in sentence:
    for key in dic:
        if key in word:
            if type(dic[key]) is list:
                for item in dic[key]:
                    word1=word.replace(key,item)
                    print(word1)
                    sentence.append(word1)
                    print(key,item,word)
            else:
                word1=word.replace(key,item)
                print("else",word1)
                sentence.append(word1)
                print(key,item,word)
            count2+=1
```
